```python
import numpy as np

def linear_regression_gd(X, Y, lr=0.08, epochs=10000):
    m = 0.0   # slope
    c = 0.0   # intercept
    n = len(X)

    for step in range(epochs):
        Y_hat = m * X + c          # predicted values
        error = Y - Y_hat          # residuals

        loss = (1/n) * np.sum(error ** 2)

        dm = (-2/n) * np.sum(X * error)
        dc = (-2/n) * np.sum(error)

        m -= lr * dm
        c -= lr * dc

        if step % 1000 == 0:
            print(f"Step {step}: m={m:.4f}, c={c:.4f}, loss={loss:.4f}")

    return m, c
```

RUN THE FUNCTION

```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([5, 7, 9, 11, 13])

m, c = linear_regression_gd(X, Y)

print("\nFinal model:")
print(f"y = {m:.2f}x + {c:.2f}")
```
